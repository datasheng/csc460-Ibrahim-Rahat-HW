```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('twitchdata-update.csv')

# print(df.head())
# df.info()
# df.describe()

histos = ['Watch time(Minutes)', 'Stream time(minutes)', 'Peak viewers', 'Average viewers', 'Followers', 
'Followers gained', 'Views gained']

# 1-dimensional histograms
for col in histos:
    sns.histplot(df[col], bins=20)
    plt.title(col)
    plt.show()

# Categorical data for top 10 (too many rows)

top10 = df.nlargest(10, 'Followers')
sns.barplot(data=top10, x='Followers', y='Channel', orient='h')
plt.show()

# Box plot
plt.figure(figsize=(12,6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.xticks(rotation=45)
plt.title("Box plot of all numerical variables")
plt.show()



```
